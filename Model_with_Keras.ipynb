 ## **A. Build a baseline model**

```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

import warnings
warnings.simplefilter('ignore', FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning, module='keras')
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head(10)
```

```python
concrete_data.shape
```

```python
concrete_data.describe()
```

```python
concrete_data.isnull().sum()
```

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
predictors.head()
```

```python
target.head()
```

```python
n_cols = predictors.shape[1] # number of predictors
n_cols
```

```python
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))

    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
x_train,x_test,y_train,y_test = train_test_split(predictors,target,test_size=0.3)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
num_models = 50
mse_list = []
for i in range(num_models):
    x_train,x_test,y_train,y_test = train_test_split(predictors,target,test_size=0.3)
    model = regression_model()
    model.fit(x_train, y_train,epochs=50, verbose=0)
    predict = model.predict(x_test)
    mse = mean_squared_error(y_test,predict)
    mse_list.append(mse)
```

```python
average_mse = np.mean(mse_list)
std_dev = np.std(mse_list)

print("mean of  MSE:", average_mse)
print("Standard deviation of MSE:", std_dev)
```

## **B. Normalize the data**

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
target_norm = (target - target.mean()) / target.std()
```

```python
predictors_norm.head()
```

```python
target_norm.head()
```

```python
n_cols = predictors_norm.shape[1]
n_cols # number of predictors
```

```python
num_models = 50
mse_list_normalized = []
for i in range(num_models):
    x_train,x_test,y_train,y_test = train_test_split(predictors_norm,target_norm,test_size=0.3)
    model = regression_model()
    model.fit(x_train, y_train,epochs=50, verbose=0)
    predict = model.predict(x_test)
    mse = mean_squared_error(y_test,predict)
    mse_list_normalized.append(mse)
```

```python
average_mse_normalized = np.mean(mse_list_normalized)
std_dev_normalized = np.std(mse_list_normalized)

print("Mean of the MSE of the standardized data", average_mse)
print("Standard deviation of the MSE of the standardized data:", std_dev)
```

How does the mean of the mean squared errors compare to that from Step A?


```python
print('The difference between two values {} and {}'.format(average_mse,average_mse_normalized))
```

## **C. Increase the number of epochs to 100**

```python
num_models = 50
mse_list_100epochs = []
for i in range(num_models):
    x_train,x_test,y_train,y_test = train_test_split(predictors_norm,target_norm,test_size=0.3)
    model = regression_model()
    model.fit(x_train, y_train,epochs=100, verbose=0)
    predict = model.predict(x_test)
    mse = mean_squared_error(y_test,predict)
    mse_list_100epochs.append(mse)
```

```python
average_mse_epochs = np.mean(mse_list_100epochs)
std_dev_epochs = np.std(mse_list_100epochs)

print("Mean of the MSE ", average_mse_epochs)
print("Standard deviation of the MSE :", std_dev_epochs)
```

```python
print('The difference between two values {} and {}'.format(average_mse_normalized,average_mse_epochs))
```

We can know that more training can make the error smaller

## **D. Increase the number of hidden layers**

```python
def morelayersmodel():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))

    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
num_models = 50
mse_list = []
for i in range(num_models):
    x_train,x_test,y_train,y_test = train_test_split(predictors_norm,target_norm,test_size=0.3)
    model = morelayersmodel()
    model.fit(x_train, y_train,epochs=50, verbose=0)
    predict = model.predict(x_test)
    mse = mean_squared_error(y_test,predict)
    mse_list.append(mse)
```

```python
average_mse = np.mean(mse_list)
std_dev = np.std(mse_list)

print("Mean of the MSE ", average_mse)
print("Standard deviation of the MSE :", std_dev)
```

```python
print('The difference between two values {} and {}'.format(average_mse_normalized,average_mse))
```

We can know that adding more layers can also make the error smaller.
